## 2. Validate AMFI Code

```python
nav_history = pd.read_csv(
    RAW_DIR / "02_nav_history.csv"
)

master_codes = set(fund_master["amfi_code"])

nav_codes = set(nav_history["amfi_code"])

print("Fund Master:", len(master_codes))
print("NAV History:", len(nav_codes))
```

```python
missing = master_codes - nav_codes

if missing:
    print(missing)
else:
    print("All AMFI codes validated successfully.")
```

```python
orphan = nav_codes - master_codes

if orphan:
    print(orphan)
else:
    print("No orphan NAV records.")
```
